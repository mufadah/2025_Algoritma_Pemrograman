#Praktikum 01 - Bubble Sort

```python
def bubble_sort(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j] > data[j + 1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

angka = [64, 34, 25, 12, 22, 11, 90]
print("Data sebelum diurutkan: ", angka)
print("Data setelah bubble sort: ", bubble_sort(angka))
```

#Praktikum 02 - Selection Sort

```python
def selection_sort(data):
  n = len(data)
  for i in range(n):
    min_index = i
    for j in range (i + 1, n):
      if data[j] < data[min_index]:
        min_index = j
    data[i], data[min_index] = data[min_index], data[i]
  return data

angka = [64, 25, 12, 22, 11]
print("Data sebelum diurutkan:", angka)
print("Data setelah Selection Sort:", selection_sort(angka))
```

#Praktikum 03 - Insertion Sort

```python
def insertion_sort(data):
  n = len(data)
  for i in range (1, n):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

angka = [12, 11, 13, 5, 6]
print("Data sebelum diurutkan:", angka)
print("Data setelah Insertion Sort:", insertion_sort(angka))
```

#Praktikum 04 - Studi Kasus Nilai Mahasiswa

```python
def bubble_sort_nilai(data):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][1] < data[j + 1][1]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

mahasiswa = [("Andi", 80), ("Budi", 70), ("Cici", 90), ("Doni", 85)]
print("Data sebelum diurutkan:", mahasiswa)
print("Data setelah diurutkan (desc):", bubble_sort_nilai(mahasiswa))
```

#Praktikum 06 - Sorting Data Barang

```python
def bubble_sort_key(data, key_index = 0, reverse = False):
  n = len(data)
  for i in range (n):
    for j in range (0, n - i - 1):
      if reverse:
        if data[j][key_index] < data[j + 1][key_index]:
          data[j], data[j + 1] = data[j + 1], data[j]
  return data

barang = [("Buku", 15000), ("Pensil", 5000), ("Tas", 75000), ("Pulpen", 10000)]
print("Urut berdasarkan nama (asc):", bubble_sort_key(barang.copy(), 0))
print("Urut berdasarkan harga (desc):", bubble_sort_key(barang.copy(), 1, reverse=True))
```

```python
def input_items():
    items = []
    while True:
        nama = input("Masukkan nama barang (kosongkan untuk selesai): ").strip()
        if not nama:
            break
        while True:
            harga = input("Masukkan harga (angka): ").strip()
            try:
                harga_val = float(harga)
                break
            except ValueError:
                print("Harga tidak valid. Masukkan angka, misalnya 15000 atau 99.99.")
        items.append({"name": nama, "price": harga_val})
    return items

def display_items(items):
    if not items:
        print("(Tidak ada data barang)\n")
        return
    print("\nDaftar barang:")
    print(f"{'No':<4}{'Nama':<30}{'Harga':>12}")
    print("-" * 46)
    for i, it in enumerate(items, start=1):
        print(f"{i:<4}{it['name']:<30}{it['price']:>12.2f}")
    print()

def sort_items(items, by="name", ascending=True):
    if by == "name":
        keyfunc = lambda it: it["name"].casefold()
    else:
        keyfunc = lambda it: it["price"]
    return sorted(items, key=keyfunc, reverse=not ascending)

def main():
    print("=== Program Manajemen Barang Sederhana ===")
    items = input_items()
    while True:
        print("Pilih aksi:")
        print("1. Tampilkan daftar barang")
        print("2. Urutkan barang")
        print("3. Tambah barang")
        print("4. Keluar")
        pilihan = input("Masukkan pilihan (1-4): ").strip()

        if pilihan == "1":
            display_items(items)

        elif pilihan == "2":
            if not items:
                print("Tidak ada barang untuk diurutkan.\n")
                continue
            print("Urutkan berdasarkan:")
            print("a. Nama (A-Z / Z-A)")
            print("b. Harga (Naik / Turun)")
            opsi = input("Pilih (a/b): ").strip().lower()
            if opsi == "a":
                arah = input("Ascending (A) atau Descending (D)? ").strip().lower()
                asc = True if arah == "a" or arah == "ascending" else False
                items = sort_items(items, by="name", ascending=asc)
                display_items(items)
            elif opsi == "b":
                arah = input("Ascending (Naik) (A) atau Descending (Turun) (D)? ").strip().lower()
                asc = True if arah == "a" or arah == "ascending" else False
                items = sort_items(items, by="price", ascending=asc)
                display_items(items)
            else:
                print("Pilihan tidak valid.\n")

        elif pilihan == "3":
            new_items = input_items()
            items.extend(new_items)
            print("Barang berhasil ditambahkan.\n")

        elif pilihan == "4":
            print("Keluar. Terima kasih.")
            break

        else:
            print("Pilihan tidak dikenal. Coba lagi.\n")

if __name__ == "__main__":
    main()
```
