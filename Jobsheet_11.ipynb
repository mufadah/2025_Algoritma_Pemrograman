#Praktikum 01 - Sequential Search pada Data Angka

```python
def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indexs {i}"
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

#Praktikum 02 - Binary Search pada Data Angka

```python
def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indexs {mid}"
    elif data[mid] < target:
      left = mid +1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

#Praktikum 03 - Sequential Search untuk Data Nama

```python
def sequential_search_nama(data, target):
  for i in range(len(data)):
    if data[i].lower() == target.lower():
      return f"Nama {target} ditemukan di indexs {i + 1}"
  return f"Nama {target} tidak ditemukan!"

nama = ["Andi", "Budi", "Cici", "Doni"]
print("Data", nama)
print(sequential_search_nama(nama, "Cici"))
print(sequential_search_nama(nama, "Agus"))
```

#Praktikum 04 - Studi Kasus Nilai Mahasiswa dengan Binary Search

```python
def binary_search_nilai(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right)//2
    if data[mid] == target:
      return f"Data {target} ditemukan di indexs {mid}"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

nilai = [60, 70, 75, 80, 85, 90]
print("Data", nilai)
print(binary_search_nilai(nilai, 85))
print(binary_search_nilai(nilai, 100))
```

#Praktikum 05 - Perbandingan Waktu Sequential vs Binary Search

```python
from time import time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan"
  return -1

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan"
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return -1

data = list(range(1,100001))
target = 99999

start = time()
sequential_search(data, target)
end = time()
print(f"Sequential search membutuhkan waktu {end - start:.5f} detik")

start = time()
binary_search(data, target)
end = time()
print(f"Binary search membutuhkan waktu {end - start:.5f} detik")
```

#Praktikum 06 - Kombinasi Sorting dan Searching

```python
from random import randint

def insertion_sort(data):
  for i in range(1, len(data)):
    key = data[i]
    j = i - 1
    while j >= 0 and data[j] > key:
      data[j + 1] = data[j]
      j -= 1
    data[j + 1] = key
  return data

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return mid
    elif data[mid] < target:
      left = mid + 1
    else:
      right = mid - 1
  return f"{target} tidak ditemukan"

data = [randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan",data)
urut = insertion_sort(data)
print(f"Data setelah diurutkan {urut}")
target = urut[7]
print(f"Binary search dari {target} ada di index {binary_search(urut, target)}")
```

#Penugasan 04 - Studi Kasus Barang Toko

```python
def bubble_sort(data, key_index=1):
  n = len(data)
  for i in range(n):
    for j in range(0, n - i - 1):
      if data[j][key_index] > data[j + 1][key_index]:
        data[j], data[j + 1] = data[j + 1], data[j]
  return data

def binary_search(data, target, key_index=1):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid][key_index] == target:
      return f"Data {target} ditemukan di indexs {mid}"
    elif data[mid][key_index] < target:
      left = mid +1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

data = [("Tas", 10000), ("Jas", 200000), ("Sandal", 13500), ("Pulpen", 5300), ("Aqua", 5000)]
print("Data sebelum dirapihkan", data)
print("\nData setelah dirapihkan", bubble_sort(data))
target = int(input("Masukkan harga barang: "))
print(binary_search(data, target))
```

#Penugasan 05 - Analisis Kompleksitas

```python
import random, time

def sequential_search(data, target):
  for i in range(len(data)):
    if data[i] == target:
      return f"Data {target} ditemukan di indexs {i}"
  return f"Data {target} tidak ditemukan"

def binary_search(data, target):
  left, right = 0, len(data)-1
  while left <= right:
    mid = (left + right) // 2
    if data[mid] == target:
      return f"Data {target} ditemukan di indexs {mid}"
    elif data[mid] < target:
      left = mid +1
    else:
      right = mid - 1
  return f"Data {target} tidak ditemukan"

def jalankan_pengujian(ukuran):
    for i in ukuran:
      data = [random.randint(1,1000) for i in range(i)]
      print(f"\n=== Ukuran data: {i} ===")
      print(data)
      for func in [sequential_search, binary_search]:
          data = data.copy()
          start = time.time()
          end = time.time()
          print(f"{func.__name__} membutuhkan waktu {end - start:.10f} detik")

ukuran = [10, 50, 100]
jalankan_pengujian(ukuran)
```
